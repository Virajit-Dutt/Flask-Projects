```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Generate dummy data
n = 100  # number of rows

# Generate date range
start_date = datetime(2023, 1, 1)
date_range = [start_date + timedelta(days=i) for i in range(n)]

# Stock tickers available
stock_tickers = ['AAPL', 'MSFT', 'TSLA', 'GOOG', 'AMZN']

# Generate random stock prices, sentiment scores, and portfolio values
stock_prices = [random.uniform(800, 1500) for _ in range(n)]
sentiment_scores = [random.uniform(-1, 1) for _ in range(n)]
sentiment_sources = random.choices(['Twitter', 'Google News', 'Reddit'], k=n)
portfolio_values = [random.uniform(90000, 100000) for _ in range(n)]

# Sectors for each stock ticker (example sectors)
stock_sectors = {
    'AAPL': 'Technology',
    'MSFT': 'Technology',
    'TSLA': 'Automotive',
    'GOOG': 'Technology',
    'AMZN': 'Retail'
}

# Generate dummy data for each stock ticker and date
data = []

for date in date_range:
    for ticker in stock_tickers:
        row = {
            'Date': date,
            'Stock Ticker': ticker,
            'Stock Price': random.uniform(100, 1500),  # Random stock price
            'Sentiment Score': random.uniform(-1, 1),  # Random sentiment score
            'Sentiment Source': random.choice(['Twitter', 'Google News', 'Reddit']),  # Random sentiment source
            'Portfolio Value': random.uniform(90000, 100000),  # Random portfolio value
            'Sector': stock_sectors[ticker]  # Sector for each stock
        }
        data.append(row)

# Convert data to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('financial_sentiment_data_with_sector.csv', index=False)

print("Dummy financial and sentiment data with sectors generated and saved to 'financial_sentiment_data_with_sector.csv'.")
```

```python
import pandas as pd
import plotly.express as px

# Load the CSV data
df = pd.read_csv('data//findata.csv')
```

```python
fig1 = px.line(df, x='Date', y='Portfolio Value', color='Stock Ticker',
               title='Portfolio Value Over Time by Stock Ticker')
fig1.update_xaxes(tickangle=45)  # Rotate x-axis labels for better visibility
fig1.show()
```

```python
fig2 = px.bar(df.groupby('Sentiment Source')['Sentiment Score'].mean().reset_index(),
              x='Sentiment Source', y='Sentiment Score', title='Average Sentiment Score by Source')
fig2.show()
```

```python
fig3 = px.pie(df, names='Stock Ticker', title='Stock Ticker Distribution')
fig3.show()
```

```python
fig4 = px.scatter(df[df['Stock Ticker']=='AAPL'], x='Stock Price', y='Sentiment Score', color='Stock Ticker',
                  title='Sentiment Score vs. Stock Price by Stock Ticker')
fig4.show()
```

```python
fig5 = px.histogram(df, x='Portfolio Value', color='Sector', title='Portfolio Value Distribution by Sector')
fig5.show()
```

```python
# Determine the size of each chunk (ensure the DataFrame can be evenly split)
num_chunks = 5
chunk_size = len(df) // num_chunks

# Split the DataFrame into 5 equal parts
dfs = [df[i*chunk_size:(i+1)*chunk_size] for i in range(num_chunks)]

# Save each part with different file extensions
dfs[0].to_csv('data//part1.csv', index=False)        # Save as .csv
dfs[1].to_excel('data//part2.xlsx', index=False)      # Save as .xlsx
dfs[2].to_json('data//part3.json', orient='records')  # Save as .json
dfs[3].to_parquet('data//part4.parquet', index=False) # Save as .parquet
dfs[4].to_html('data//part5.html', index=False)       # Save as .html

print("DataFrames saved with different extensions.")
```
